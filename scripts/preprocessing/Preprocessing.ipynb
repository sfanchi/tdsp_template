En terminos generaes, tenemos un df bastante limpio y listo para procesar, sin embargo para mayor claridad en nuestros resultados, podemos aplicar una funcion de preprocesamiento de texto a nuestra columna de 'NOMBRE_CORTO_DE_EXPERIENCIA' para asi tener los textos sin errores:

```python
import re
from unidecode import unidecode

def clean_text(text):
    """
    Argumento:
        text (str): El texto a ser limpiado.

    Devuelve:
        str: El texto limpio.
    """
    # quitamos las URL si existen
    text = re.sub(r'http\S+', '', text)

    # quitamos tags de html
    text = re.sub(r'<.*?>', '', text)

    # quitamos los caracteres no alfabéticos y puntuación
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # convertimos a minusculas
    text = text.lower()

    # quitamos caracteres que no sean minusculas
    text = re.sub(r'[^a-z ]',' ', text)

    # quitamos espacios duplicados
    text = re.sub(r'\s{2,}', ' ', text)

    # quitamos acentos
    text = unidecode(text)

    return text
```

```python
df['NOMBRE_CORTO_DE_EXPERIENCIA'].apply(clean_text)
```
